```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = './BindingDB_All.tsv'

# Load the first 10,000 lines of the TSV file
data = pd.read_csv(file_path, sep='\t', nrows=1000000)



different_targets = data.groupby('Target Source Organism According to Curator or DataSource')['Target Name'].agg(['count'])
different_targets
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
```

```python
mols_df = pd.DataFrame(data['Ligand SMILES'])
print(mols_df)
```

```python
# Loop over each row in the DataFrame
for i, row in mols_df.iterrows():
    # Convert the SMILES string to a molecule object
    mol = Chem.MolFromSmiles(row['Ligand SMILES'])
    if i%100000 == 0: #show every 10000 molecule
        if mol:
            # Visualize the molecule using RDKit's drawing function
            img = Draw.MolToImage(mol)
            
            # Display the image using Matplotlib
            plt.figure(figsize=(3, 3))
            plt.imshow(img)
            plt.axis('off')  # Hide the axes
            plt.title(f"Molecule {i+1}")
            plt.show()
        
    
```

```python
keywords = ['niversi', 'ational', 'olleg', 'nstitu', 'cademy', 'ospital', 'acult']

# Get unique institutions
unique_institutions = data['Institution'].unique()

# Print the shape
print(unique_institutions.shape)

# Separate into two lists
academic_institutions = []
industry_institutions = []

for institution in unique_institutions:
   # institution = institution.lower()
    if any(keyword in institution for keyword in keywords):
        academic_institutions.append(institution)
    else:
        industry_institutions.append(institution)

# Output the sorted lists
print(len(academic_institutions), 'academic')
print(len(industry_institutions), 'industry')

# print("Academic Institutions:", academic_institutions)
# print("Industry Institutions:", industry_institutions)


```

```python
unique_organisms = data['Target Source Organism According to Curator or DataSource'].unique()

bacteria = []
virus = []
mammals = []
others = [] #include parasites, plants, yeast, frogs, ...

# Define sets of keywords for each category
bacteria_keywords = {'acinetobacter baumannii', 'zymomonas mobilis', 'leptospira interrogans','borrelia burgdorferi', 'vibrio harveyi', 'leuconostoc mesenteroides', 'yersinia enterocolitica', 'enterobacter cloacae', 'bacteroides', 'francisella tularensis','neisseria meningitidis', 'bacteroides thetaiotaomicron', 'helicobacter pylori', 'proteus vulgaris', 'aquifex aeolicus', 'clostridium', 'pestis','histolyticum', 'escherichia','bacterium', 'bacillus', 'coccus', 'staphylococcus', 'streptococcus', 'pneumoniae', 'pseudomonas', 'haemophilus', 'strain', 'serratia'}
virus_keywords = {'virus', 'retrovirus', 'bacteriophage', 'phage'}
mammal_keywords = {'bison bison', 'cricetulus griseus', 'gerbillus sp.', 'canis familiaris', 'macaca mulatta', 'Cricetidae sp.', 'mesocricetus auratus','mus', 'rattus', 'homo', 'taurus', 'oryctolagus', 'gallus', 'equus', 'ovis', 'scrofa', 'lupus', 'cavia'}

# Categorizing organisms
for organism in unique_organisms:
    if isinstance(organism, str):  # Check if organism is a string
        organism_lower = organism.lower()  # Make comparison case insensitive
        
        if any(keyword in organism_lower for keyword in virus_keywords):
            virus.append(organism)
        elif any(keyword in organism_lower for keyword in bacteria_keywords):
            bacteria.append(organism)
        elif any(keyword in organism_lower for keyword in mammal_keywords):
            mammals.append(organism)
        else:
            others.append(organism)
    else:
        others.append(organism) 

print(len(bacteria), 'bacteria')
print(len(virus), 'virus')
print(len(mammals), 'mammals')
print(len(others), 'others')

```

I classified the organisms in 4 main categories : bacteria, virus, mammals and others (yeast, parasites, fish, drosophila, frog)
